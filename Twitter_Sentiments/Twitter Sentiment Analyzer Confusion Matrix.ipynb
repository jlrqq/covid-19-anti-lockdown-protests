```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sentiment_jan_mar = pd.read_csv('./Twitter_Jan_Mar_Sentiments.csv')
sentiment_jan_mar
```

```python
sentiment_mar = pd.read_csv('./Twitter_Mar_Sentiments.csv')
sentiment_mar
```

```python
sentiment_may_nov = pd.read_csv('./Twitter_May_Nov_Sentiments.csv')
sentiment_may_nov
```

```python
sentiments_overtime = pd.concat([sentiment_jan_mar, sentiment_mar, sentiment_may_nov], axis=0)
sentiments_overtime
```

```python
# random split train and test data
index = sentiments_overtime.index
sentiments_overtime['random_number'] = np.random.randn(len(index))
train = sentiments_overtime[sentiments_overtime['random_number'] <= 0.8]
test = sentiments_overtime[sentiments_overtime['random_number'] > 0.8]
```

```python
# count vectorizer:
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
train_matrix = vectorizer.fit_transform(train['processed_text'].values.astype('U'))
test_matrix = vectorizer.transform(test['processed_text'].values.astype('U'))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
# Split target and independent variables
X_train = train_matrix
X_test = test_matrix
y_train = train['sentiment_label']
y_test = test['sentiment_label']
```

```python
# Fit model on data
lr.fit(X_train, y_train)
```

```python
# Make predictions
predictions = lr.predict(X_test)
```

```python
# Test - find accuracy, precision, recall:
from sklearn.metrics import confusion_matrix, classification_report
new = np.asarray(y_test)
cf_matrix = confusion_matrix(predictions, y_test, labels=['negative', 'neutral', 'positive'])
print(classification_report(predictions, y_test))
```

```python
# Visualize Confusion Matrix
plt.figure(figsize=(10,6))
fx=sns.heatmap(confusion_matrix(y_test, predictions), annot=True, fmt=".2f",cmap="GnBu")
fx.set_title('Confusion Matrix \n');
fx.set_xlabel('\n Predicted Values\n')
fx.set_ylabel('Actual Values\n');
fx.xaxis.set_ticklabels(['negative', 'neutral', 'positive'])
fx.yaxis.set_ticklabels(['negative', 'neutral', 'positive'])
plt.savefig('../Twitter_Output/Twitter_Confusion_Matrix.png')
plt.show()
```
