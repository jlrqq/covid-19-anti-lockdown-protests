```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment_jan_mar = pd.read_csv('./Twitter_Jan_Mar_Sentiments.csv')
sentiment_jan_mar
```

```python
# Convert the date column to datetime values
sentiment_jan_mar['Datetime'] = pd.to_datetime(sentiment_jan_mar['Datetime'])

# Make date the index of the DataFrame
sentiment_jan_mar = sentiment_jan_mar.set_index('Datetime')

sentiment_jan_mar.resample('D')['sentiment_score'].mean().plot(
    title="COVID19 Lockdown Tweet Sentiment by Days from Jan to Mar");

plt.xlabel('Date Time')
plt.ylabel('Sentiment Score')
plt.savefig('../Twitter_Output/Twitter_Jan_Mar_SentimentOvertime.png', dpi=400, bbox_inches="tight")
```

```python
sentiment_mar = pd.read_csv('./Twitter_Mar_Sentiments.csv')
sentiment_mar
```

```python
# Convert the date column to datetime values
sentiment_mar['Datetime'] = pd.to_datetime(sentiment_mar['Datetime'])

# Make date the index of the DataFrame
sentiment_mar = sentiment_mar.set_index('Datetime')

sentiment_mar.resample('D')['sentiment_score'].mean().plot(
    title="COVID19 Lockdown Tweet Sentiment by Days in Mar");

plt.xlim(left=16)
plt.ylim(bottom=10)
plt.xlabel('Date Time')
plt.ylabel('Sentiment Score')
plt.savefig('../Twitter_Output/Twitter_Mar_SentimentOvertime.png', dpi=400, bbox_inches="tight")
```

```python
sentiment_may_nov = pd.read_csv('./Twitter_May_Nov_Sentiments.csv')
sentiment_may_nov
```

```python
# Convert the date column to datetime values
sentiment_may_nov['Datetime'] = pd.to_datetime(sentiment_may_nov['Datetime'])

# Make date the index of the DataFrame
sentiment_may_nov = sentiment_may_nov.set_index('Datetime')

sentiment_may_nov.resample('D')['sentiment_score'].mean().plot(
    title="COVID19 Lockdown Tweet Sentiment by Days from May to Nov");

plt.xlabel('Date Time')
plt.ylabel('Sentiment Score')
plt.savefig('../Twitter_Output/Twitter_May_Nov_SentimentOvertime.png', dpi=400, bbox_inches="tight")
```

```python
sentiments_overtime = pd.concat([sentiment_jan_mar, sentiment_mar, sentiment_may_nov], axis=0)
```

```python
sentiments_overtime.resample('M')['sentiment_score'].mean().plot(
    title="COVID19 Lockdown Tweet Sentiment by Month");

plt.xlabel('Date Time')
plt.ylabel('Sentiment Score')
plt.savefig('../Twitter_Output/Twitter_SentimentOvertime.png', dpi=400, bbox_inches="tight")
```
