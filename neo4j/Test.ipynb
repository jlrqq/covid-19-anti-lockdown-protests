```python
import pandas as pd
import numpy as np
import re
```

```python
tweets = pd.read_csv('../Twitter_Data/Twitter_Covid-19_Lockdown_5000.csv')
```

```python
tweets.head()
```

```python
tweets.info()
```

```python
tweets['rt_username'] = np.nan
tweets['tag_usernames'] = np.nan
```

```python
for i, row in tweets.iterrows():
    # retweets
    regex_rt = r'RT[ ]*(@[A-Za-z0-9_]+)'
    match = re.search(regex_rt, row['text'])
    if match:
        sub_string = match.group()
        tweets.loc[i, 'rt_username'] = sub_string[4:]
    else:
        # tags
        regex_tags = r'@[A-Za-z0-9_]+'
        matches = re.findall(regex_tags, row['text'])
        if matches:
            tweets.at[i, 'tag_usernames'] = [s[1:] for s in matches]
```
